## Implied volatility (Jin et al. 2012)



* Volatility skeness

    Volatility of OTM put - volatility of ATM call
    Base window = [-50,-11] before event 
    - Screening criteria
        - Options with greater-than-zero open interest
        - Options with experation date between 10 and 60 days away
        - Call with delta within [0.4, 0.7] **closest to 0.5**
        - Put with delta within [-0.45,-0.15] **closest to -0.3**


* Volatility spread

    Difference in implied volatilities between call and put pairs matched strike price and maturity dates weigthed averaged by the open interest

    * Options with experation date between 10 and 60 days away



## Volume based

* Option-to-stock volume ratio (Johnson & So 2012)
    
    Sum of volumes all strike expiring within 30 trading days divdied by equity 
    volume (round lot of 100)
    * Screening criteria
        * Firm-weeks with at least 25 calls and 25 puts traded
        * At least 6 months of past weekly options
        * Exclude closed-end funds, real estate investment trusts, American depository receipts, and firms with a stock price below $1

* Open new positions vs close previous positions (Ge et al. 2016)

    Cannot be constructed from OptionMetrics, requires data from the [International Securities Exchange (ISE)](https://www.nasdaq.com/solutions/nasdaq-ise)
